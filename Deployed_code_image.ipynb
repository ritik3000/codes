```
import cv2
import imutils

def isrgb(imgpath):
    img = imutils.url_to_image(imgpath)
    if len(img.shape) < 3: return False
    if img.shape[2]  == 1: return False
    b,g,r = img[:,:,0], img[:,:,1], img[:,:,2]
    if (b==g).all() and (b==r).all(): return False
    return True
```

```
def change_gray(req):
  image=imutils.url_to_image(req)
  #image = cv2.imread(img)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  gray_three = cv2.merge([gray,gray,gray])
  return gray_three
```

```
def get_model():
  img2vec = Img2Vec(cuda=False)
  return img2vec
```

```
import json
```

```
from PIL import Image
def predict_batch(data):
  result=[]
  model=get_model()
  for i in range(0,len(data)):
    y=data[i]['image_url_1']
    print(y)
    z=urllib.request.urlopen(y)
    if(isrgb(y)):
      image_1 = Image.open(z)
      tensor_image=model.get_vec(image_1, tensor=True)
      tensor_image=tensor_image.reshape(512)
      tensor_image= tensor_image.numpy()[:]
      np_array_to_list = tensor_image.tolist()
      result.append({'image_url_1':json.dumps(np_array_to_list)})
    else:
      changed_img=change_gray(y)
      image_1=Image.fromarray(changed_img)
      tensor_image=model.get_vec(image_1, tensor=True)
      tensor_image=tensor_image.reshape(512)
      np_array_to_list = tensor_image.tolist()
      result.append({'image_url_1':json.dumps(np_array_to_list)})

  return result

    
 
```

```
ans=predict_batch(data)
```
